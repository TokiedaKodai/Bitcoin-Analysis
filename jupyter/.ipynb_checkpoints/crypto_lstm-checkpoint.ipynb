```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import sys

from keras.callbacks import CSVLogger, ModelCheckpoint
from tensorflow.python.client import device_lib

sys.path.append('../')

import lstm

print(device_lib.list_local_devices()) 
```

```python
dir_data = '../../Data/CryptoData/'
file_train = dir_data + 'train.csv'
file_hash = dir_data + 'HashRate.csv'
file_elect = dir_data + 'Electricity.csv'
file_trend = dir_data + 'Trend.csv'
file_spx = dir_data + 'SPXUSD.csv'

dir_model = '../../Model/'
```

```python
def plot_results_multiple(predicted_data, true_data, prediction_len):
    fig = plt.figure(facecolor='white')
    ax = fig.add_subplot(111)
    ax.plot(true_data, label='True Data')
    for i, data in enumerate(predicted_data):
        padding = [None for p in range(i * prediction_len)]
        plt.plot(padding + data, label='Prediction')
#         plt.legend()
    plt.show()
```

```python
df_train = pd.read_csv(file_train)
```

```python
date = df_train['Date'].values
```

```python
btc = df_train['BTCUSD_Close_log'].values
```

```python
len_data = len(btc)
```

```python
len_day = 1440
len_hour = 60
seq_len = len_hour * 1
train_len = len_day * 1

X_train, y_train, X_test, y_test = lstm.gen_data(btc[-train_len:], seq_len, normalise_window=False)
```

```python
print('Train length = ', X_train.shape[0])
print('Test length = ', X_test.shape[0])
```

```python
model_save_cb = ModelCheckpoint(dir_model + 'model-lstm_btc.hdf5',
                                    period=1,
                                    save_weights_only=True)
csv_logger_cb = CSVLogger(dir_model + 'model-lstm_btc.log', append=False)
```

```python
model = lstm.build_model([1, seq_len, seq_len*2, 1])
```

```python
epoch = 100
model.fit(X_train, 
          y_train, 
          batch_size=512, 
          epochs=epoch, 
          validation_split=0.1, 
          callbacks=[model_save_cb, csv_logger_cb],
          verbose=1)
```

```python
df_log = pd.read_csv(dir_model + 'model-lstm_btc.log')
```

```python
df_log[['loss', 'val_loss']].plot(subplots=False, figsize=(15, 5))
```

```python
pred_len = 50
predictions = lstm.predict_sequences_multiple(model, X_test, seq_len, pred_len)
```

```python
plot_results_multiple(predictions, y_test, pred_len)
```
